```python
import yfinance as yf
import pandas as pd

equities = ["AAPL", "MSFT", "GOOGL", "AMZN", "TSLA", "NVDA", "META", "JPM", "V", "DIS"]

data = yf.download(equities, start="2024-01-01", end="2025-01-01", group_by="ticker")

data.head()
```

```python
def tidy_prices(data, tickers):
    frames = []
    for ticker in tickers:
        df = data[ticker].copy()
        df = df.reset_index()
        df["ticker"] = ticker
        df = df.rename(columns={
            "Date": "price_date",
            "Open": "open_price",
            "High": "high_price",
            "Low": "low_price",
            "Close": "close_price",
            "Volume": "volume"
        })
        frames.append(df[["ticker", "price_date", "open_price", "high_price", "low_price", "close_price", "volume"]])
    return pd.concat(frames, ignore_index=True)

equities_tidy = tidy_prices(data, equities)
equities_tidy.head()
```

```python
crypto = ["BTC-USD", "ETH-USD", "SOL-USD", "ADA-USD", "XRP-USD", "DOGE-USD", "AVAX-USD", "DOT-USD", "LTC-USD", "LINK-USD"]
fx = ["EURUSD=X", "GBPUSD=X", "USDJPY=X", "AUDUSD=X", "USDCAD=X", "USDCHF=X", "NZDUSD=X", "USDCNY=X", "USDMXN=X", "USDZAR=X"]
bonds = ["TLT", "IEF", "SHY", "BND", "AGG", "LQD", "HYG", "TIP", "GOVT", "MUB"]

crypto_data = yf.download(crypto, start="2024-01-01", end="2025-01-01", group_by="ticker")
fx_data = yf.download(fx, start="2024-01-01", end="2025-01-01", group_by="ticker")
bonds_data = yf.download(bonds, start="2024-01-01", end="2025-01-01", group_by="ticker")

crypto_tidy = tidy_prices(crypto_data, crypto)
fx_tidy = tidy_prices(fx_data, fx)
bonds_tidy = tidy_prices(bonds_data, bonds)

crypto_tidy.head()
```

```python
equities_tidy["asset_type"] = "equity"
crypto_tidy["asset_type"] = "crypto"
fx_tidy["asset_type"] = "fx"
bonds_tidy["asset_type"] = "bond"

all_prices = pd.concat([equities_tidy, crypto_tidy, fx_tidy, bonds_tidy], ignore_index=True)

all_prices.shape
```

```python
from sqlalchemy import create_engine

engine = create_engine("postgresql+psycopg2://admin:password123@localhost:5432/financial_warehouse")

# Quick test
with engine.connect() as conn:
    print("Connected successfully")
```

```python
# Build a unique list of all tickers with their asset_type
asset_records = []
for ticker in equities:
    asset_records.append({"ticker": ticker, "asset_type": "equity", "currency": "USD"})
for ticker in crypto:
    asset_records.append({"ticker": ticker, "asset_type": "crypto", "currency": "USD"})
for ticker in fx:
    asset_records.append({"ticker": ticker, "asset_type": "fx", "currency": "USD"})
for ticker in bonds:
    asset_records.append({"ticker": ticker, "asset_type": "bond", "currency": "USD"})

assets_df = pd.DataFrame(asset_records)
assets_df.head()
```

```python
assets_df.to_sql("assets", engine, if_exists="append", index=False)
print("Assets inserted")
```

```python
asset_map = pd.read_sql("SELECT asset_id, ticker FROM assets", engine)
asset_map.head()
```

```python
prices_to_load = all_prices.merge(asset_map, on="ticker", how="left")

# Check nothing failed to map
print(prices_to_load["asset_id"].isna().sum(), "unmapped rows")

prices_to_load = prices_to_load[["asset_id", "price_date", "open_price", "high_price", "low_price", "close_price", "volume"]]
prices_to_load.head()
```

```python
prices_to_load.to_sql("prices", engine, if_exists="append", index=False)
print(f"Inserted {len(prices_to_load)} price rows")
```

```python
import numpy as np

np.random.seed(42)  # reproducible results

def generate_trades(prices_df, asset_map, trades_per_asset=8):
    trades = []
    for asset_id in prices_df["asset_id"].unique():
        asset_prices = prices_df[prices_df["asset_id"] == asset_id].sort_values("price_date").reset_index(drop=True)
        n = len(asset_prices)
        if n < trades_per_asset:
            continue
        
        # Pick random trade dates spread across the year
        trade_indices = sorted(np.random.choice(n, size=trades_per_asset, replace=False))
        
        position = 0
        for idx in trade_indices:
            row = asset_prices.iloc[idx]
            # First trade must be a BUY (can't sell what you don't own)
            if position == 0:
                side = "BUY"
            else:
                side = np.random.choice(["BUY", "SELL"], p=[0.6, 0.4])
            
            quantity = round(np.random.uniform(5, 50), 2)
            price = row["close_price"]
            fees = round(price * quantity * 0.001, 2)  # 0.1% fee
            
            if side == "BUY":
                position += quantity
            else:
                quantity = min(quantity, position)  # can't sell more than held
                position -= quantity
            
            if quantity > 0:
                trades.append({
                    "asset_id": asset_id,
                    "trade_date": row["price_date"],
                    "side": side,
                    "quantity": quantity,
                    "price": price,
                    "fees": fees
                })
    return pd.DataFrame(trades)

trades_df = generate_trades(prices_to_load, asset_map)
trades_df.shape
```

```python
trades_df.to_sql("trades", engine, if_exists="append", index=False)
print(f"Inserted {len(trades_df)} trades")
```

```python
from reportlab.lib.pagesizes import A4
from reportlab.lib import colors
from reportlab.lib.styles import getSampleStyleSheet
from reportlab.lib.units import cm
from reportlab.platypus import SimpleDocTemplate, Table, TableStyle, Paragraph, Spacer
import pandas as pd
from sqlalchemy import create_engine
from datetime import datetime

engine = create_engine("postgresql+psycopg2://admin:password123@localhost:5432/financial_warehouse")

print("Ready to build report")
```

```python
# Exposure by asset class
exposure_df = pd.read_sql("""
    WITH positions AS (
        SELECT asset_id,
            SUM(CASE WHEN side = 'BUY' THEN quantity ELSE -quantity END) AS net_position
        FROM trades GROUP BY asset_id
    ),
    latest_price AS (
        SELECT DISTINCT ON (asset_id) asset_id, close_price AS current_price
        FROM prices ORDER BY asset_id, price_date DESC
    )
    SELECT a.asset_type, a.ticker,
        ROUND(p.net_position, 2) AS net_position,
        ROUND(lp.current_price, 4) AS current_price,
        ROUND(p.net_position * lp.current_price, 2) AS exposure_usd
    FROM positions p
    JOIN assets a ON p.asset_id = a.asset_id
    JOIN latest_price lp ON p.asset_id = lp.asset_id
    WHERE p.net_position > 0
    ORDER BY a.asset_type, exposure_usd DESC
""", engine)

# Unrealized PnL
unrealized_df = pd.read_sql("""
    WITH avg_cost AS (
        SELECT asset_id, SUM(quantity * price) / SUM(quantity) AS avg_buy_price
        FROM trades WHERE side = 'BUY' GROUP BY asset_id
    ),
    positions AS (
        SELECT asset_id,
            SUM(CASE WHEN side = 'BUY' THEN quantity ELSE -quantity END) AS net_position
        FROM trades GROUP BY asset_id
    ),
    latest_price AS (
        SELECT DISTINCT ON (asset_id) asset_id, close_price AS current_price
        FROM prices ORDER BY asset_id, price_date DESC
    )
    SELECT a.ticker, a.asset_type,
        ROUND(p.net_position, 2) AS net_position,
        ROUND(ac.avg_buy_price, 4) AS avg_buy_price,
        ROUND(lp.current_price, 4) AS current_price,
        ROUND(p.net_position * (lp.current_price - ac.avg_buy_price), 2) AS unrealized_pnl
    FROM positions p
    JOIN assets a ON p.asset_id = a.asset_id
    JOIN avg_cost ac ON p.asset_id = ac.asset_id
    JOIN latest_price lp ON p.asset_id = lp.asset_id
    WHERE p.net_position > 0
    ORDER BY unrealized_pnl DESC
""", engine)

# Monthly performance
performance_df = pd.read_sql("""
    WITH positions_at_date AS (
        SELECT p.price_date, p.asset_id, p.close_price,
            COALESCE((
                SELECT SUM(CASE WHEN t.side = 'BUY' THEN t.quantity ELSE -t.quantity END)
                FROM trades t
                WHERE t.asset_id = p.asset_id AND t.trade_date <= p.price_date
            ), 0) AS position_at_date
        FROM prices p
    ),
    daily_portfolio AS (
        SELECT price_date, SUM(position_at_date * close_price) AS portfolio_value
        FROM positions_at_date WHERE position_at_date > 0
        GROUP BY price_date
    )
    SELECT DATE_TRUNC('month', price_date) AS month,
        ROUND(AVG(portfolio_value), 2) AS avg_monthly_value,
        ROUND(MAX(portfolio_value), 2) AS peak_value,
        ROUND(MIN(portfolio_value), 2) AS trough_value
    FROM daily_portfolio
    GROUP BY DATE_TRUNC('month', price_date)
    ORDER BY month
""", engine)

# Summary stats
total_exposure = exposure_df["exposure_usd"].sum()
total_unrealized = unrealized_df["unrealized_pnl"].sum()
num_positions = len(exposure_df)

print(f"Total Exposure: ${total_exposure:,.2f}")
print(f"Total Unrealized PnL: ${total_unrealized:,.2f}")
print(f"Open Positions: {num_positions}")
```

```python
doc = SimpleDocTemplate(
    "financial_report.pdf",
    pagesize=A4,
    rightMargin=2*cm, leftMargin=2*cm,
    topMargin=2*cm, bottomMargin=2*cm
)

styles = getSampleStyleSheet()
elements = []

# ── Title ──────────────────────────────────────────────
elements.append(Paragraph("Financial Data Warehouse Report", styles["Title"]))
elements.append(Paragraph(f"Generated: {datetime.now().strftime('%Y-%m-%d %H:%M')}", styles["Normal"]))
elements.append(Spacer(1, 0.5*cm))

# ── Portfolio Summary ───────────────────────────────────
elements.append(Paragraph("Portfolio Summary", styles["Heading1"]))
summary_data = [
    ["Metric", "Value"],
    ["Total Exposure (USD)", f"${total_exposure:,.2f}"],
    ["Total Unrealized PnL (USD)", f"${total_unrealized:,.2f}"],
    ["Open Positions", str(num_positions)],
    ["Asset Classes", "Equity, Crypto, FX, Bond"],
    ["Data Period", "2024-01-01 to 2025-01-01"],
]
summary_table = Table(summary_data, colWidths=[8*cm, 8*cm])
summary_table.setStyle(TableStyle([
    ("BACKGROUND", (0,0), (-1,0), colors.darkblue),
    ("TEXTCOLOR", (0,0), (-1,0), colors.white),
    ("FONTNAME", (0,0), (-1,0), "Helvetica-Bold"),
    ("ALIGN", (0,0), (-1,-1), "CENTER"),
    ("GRID", (0,0), (-1,-1), 0.5, colors.grey),
    ("ROWBACKGROUNDS", (0,1), (-1,-1), [colors.whitesmoke, colors.lightgrey]),
]))
elements.append(summary_table)
elements.append(Spacer(1, 0.5*cm))

# ── Exposure by Asset Class ─────────────────────────────
elements.append(Paragraph("Exposure by Asset Class", styles["Heading1"]))
exposure_summary = exposure_df.groupby("asset_type")["exposure_usd"].sum().reset_index()
exposure_summary.columns = ["Asset Type", "Total Exposure (USD)"]
exposure_summary["Total Exposure (USD)"] = exposure_summary["Total Exposure (USD)"].apply(lambda x: f"${x:,.2f}")
exp_data = [exposure_summary.columns.tolist()] + exposure_summary.values.tolist()
exp_table = Table(exp_data, colWidths=[8*cm, 8*cm])
exp_table.setStyle(TableStyle([
    ("BACKGROUND", (0,0), (-1,0), colors.darkblue),
    ("TEXTCOLOR", (0,0), (-1,0), colors.white),
    ("FONTNAME", (0,0), (-1,0), "Helvetica-Bold"),
    ("ALIGN", (0,0), (-1,-1), "CENTER"),
    ("GRID", (0,0), (-1,-1), 0.5, colors.grey),
    ("ROWBACKGROUNDS", (0,1), (-1,-1), [colors.whitesmoke, colors.lightgrey]),
]))
elements.append(exp_table)
elements.append(Spacer(1, 0.5*cm))

# ── Top 10 Unrealized PnL ───────────────────────────────
elements.append(Paragraph("Top 10 Positions by Unrealized PnL", styles["Heading1"]))
top10 = unrealized_df.head(10)[["ticker", "asset_type", "net_position", "avg_buy_price", "current_price", "unrealized_pnl"]].copy()
top10.columns = ["Ticker", "Type", "Position", "Avg Cost", "Current Price", "Unrealized PnL"]
top10["Unrealized PnL"] = top10["Unrealized PnL"].apply(lambda x: f"${x:,.2f}")
top10["Current Price"] = top10["Current Price"].apply(lambda x: f"${x:,.4f}")
top10["Avg Cost"] = top10["Avg Cost"].apply(lambda x: f"${x:,.4f}")
pnl_data = [top10.columns.tolist()] + top10.values.tolist()
pnl_table = Table(pnl_data, colWidths=[3*cm, 2.5*cm, 2.5*cm, 3*cm, 3*cm, 3*cm])
pnl_table.setStyle(TableStyle([
    ("BACKGROUND", (0,0), (-1,0), colors.darkblue),
    ("TEXTCOLOR", (0,0), (-1,0), colors.white),
    ("FONTNAME", (0,0), (-1,0), "Helvetica-Bold"),
    ("ALIGN", (0,0), (-1,-1), "CENTER"),
    ("GRID", (0,0), (-1,-1), 0.5, colors.grey),
    ("ROWBACKGROUNDS", (0,1), (-1,-1), [colors.whitesmoke, colors.lightgrey]),
    ("FONTSIZE", (0,0), (-1,-1), 8),
]))
elements.append(pnl_table)
elements.append(Spacer(1, 0.5*cm))

# ── Monthly Performance ─────────────────────────────────
elements.append(Paragraph("Monthly Portfolio Performance", styles["Heading1"]))
perf = performance_df.copy()
perf["month"] = perf["month"].dt.strftime("%Y-%m")
perf.columns = ["Month", "Avg Value", "Peak Value", "Trough Value"]
for col in ["Avg Value", "Peak Value", "Trough Value"]:
    perf[col] = perf[col].apply(lambda x: f"${x:,.2f}")
perf_data = [perf.columns.tolist()] + perf.values.tolist()
perf_table = Table(perf_data, colWidths=[4*cm, 5*cm, 5*cm, 5*cm])
perf_table.setStyle(TableStyle([
    ("BACKGROUND", (0,0), (-1,0), colors.darkblue),
    ("TEXTCOLOR", (0,0), (-1,0), colors.white),
    ("FONTNAME", (0,0), (-1,0), "Helvetica-Bold"),
    ("ALIGN", (0,0), (-1,-1), "CENTER"),
    ("GRID", (0,0), (-1,-1), 0.5, colors.grey),
    ("ROWBACKGROUNDS", (0,1), (-1,-1), [colors.whitesmoke, colors.lightgrey]),
    ("FONTSIZE", (0,0), (-1,-1), 8),
]))
elements.append(perf_table)

# ── Build PDF ───────────────────────────────────────────
doc.build(elements)
print("Report saved as financial_report.pdf")
```
